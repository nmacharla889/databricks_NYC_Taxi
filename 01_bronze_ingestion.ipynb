# Bronze Layer — Raw Ingestion
## Purpose
Ingest raw NYC Yellow Taxi trip data (January 2026) from the source volume 
into a Delta table without any transformations.
In a production pipeline this data would arrive automatically from cloud 
storage (Azure Blob / S3) via an ADF or Airflow ingestion job.

```python
df_raw = spark.read.parquet("/Volumes/nyc/default/nyc_raw_data/yellow_tripdata_2026-01.parquet")
```

```python
print(f"Row Count: {df_raw.count()}")
```

```python
print(f"Columns: {df_raw.columns}")
```

## Write to Bronze Delta Table
Write the raw data as-is into a Delta table.
No transformations at this layer — Bronze is the raw record of what arrived.

```python
df_raw.write.format("delta").mode("overwrite").saveAsTable("nyc.default.bronze_yellow_taxi")
display(df_raw)
```

## Verify Bronze Table
Confirm row count and schema match the raw source file.

```python
df_bronze=spark.read.table("nyc.default.bronze_yellow_taxi")
print(f"Bronze row count: {df_bronze.count()}")
display(df_bronze.limit(5))
```

## Delta History
Delta automatically tracks every write operation.
This is the foundation of time travel — every version is queryable.

```python
display(spark.sql("DESCRIBE HISTORY nyc.default.bronze_yellow_taxi"))
```
